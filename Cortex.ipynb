```python
df = df.loc[df['cidade'] == 'Rio de Janeiro']
```

```python
df['cidade'].unique()
```

```python
X= df.drop(['faturamento', 'potencial', 'nome', 'cidade', 'estado'], axis=1) 
```

```python
y = df['faturamento']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Dividir a amostra em treino (70%) e teste (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar e treinar o modelo
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Fazer previsões no conjunto de teste
y_pred = model.predict(X_test)

# Avaliar o modelo
mse = mean_squared_error(y_test, y_pred)
print(f"MSE no teste: {mse:.4f}")
```

```python
import statsmodels.api as sm

# Adicionar constante (termo de interceptação)
X = sm.add_constant(X)

# Ajustar o modelo com toda a amostra
model = sm.OLS(y, X).fit()

# Exibir os resultados
print(model.summary())
```
